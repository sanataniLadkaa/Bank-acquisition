## Interest rate of 2.9% for each customer for any due amount

```python
import pandas as pd
for i in range(1,13):
  dfjan20044 = df2004.loc[df2004['Month'].dt.month==i]
  jk =dfjan20044["Costomer"]
  jk =jk.T.drop_duplicates().T

  for j in jk.index.values:
    dfjan20044 = dfjan20044[dfjan20044['Costomer']== str(jk[j])]
    jk.T.drop_duplicates().T

    if dfjan20044['Repayment'].sum() > dfjan20044['Spend'].sum():
      print((dfjan20044['Repayment'].sum()-dfjan20044['Spend'].sum())*0.29 , "in",i,"th month of 2004\n\n")
    else:
      print( str(jk[j])," will have to give ", dfjan20044['Repayment'].sum(),"in ",i,"th month of 2004\n\n")

```
